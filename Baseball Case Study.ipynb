```python
x= df.drop(['W'], axis=1)
y = df['W']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=20)
print("Shape of xtrain:",xtrain.shape)
print("Shape of xtest:",xtest.shape)
print("Shape of ytrain:",ytrain.shape)
print("Shape of ytest:",ytest.shape)
```

```python
# feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(xtrain)
X_test_scaled = scaler.transform(xtest)
```

```python
# linear regression
lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled, ytrain)
lin_reg_pred = lin_reg.predict(X_test_scaled)
lin_reg_rmse = mean_squared_error(ytest, lin_reg_pred, squared=False)

print('Linear Regression RMSE:', lin_reg_rmse)

# decision tree regression
tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train_scaled, ytrain)
tree_reg_pred = tree_reg.predict(X_test_scaled)
tree_reg_rmse = mean_squared_error(ytest, tree_reg_pred, squared=False)
print('Decision Tree Regression RMSE:', tree_reg_rmse)

# random forest regression
forest_reg = RandomForestRegressor()
forest_reg.fit(X_train_scaled, ytrain)
forest_reg_pred = forest_reg.predict(X_test_scaled)
forest_reg_rmse = mean_squared_error(ytest, forest_reg_pred, squared=False)
print('Random Forest Regression RMSE:', forest_reg_rmse)

svc=SVC()
svc.fit(X_train_scaled, ytrain)
svc_pred = forest_reg.predict(X_test_scaled)
svc_rmse = mean_squared_error(ytest, forest_reg_pred, squared=False)
print('Support Vector RMSE:', forest_reg_rmse)

# select the best model based on its performance on the testing set
best_model = forest_reg
```
